# Q4. What is the role of weights and biases in forward propagation?

There is a very impotant role of weights and biases in the forward propagation.

**Weights**

Each connection between nodes (neurons) in the network has an associated weight. The weight represents the strength of the connection and signifies the importance of the corresponding input.

**Biases**

Biases are additional parameters in each neuron that allow the network to make predictions even when all input features are zero.Biases help the model account for certain offsets or biases in the data that might not be captured by the weights alone.

in a equation = z = x1.w1+x2.w2+x3.w3....+b
- w are weights
- b is bias

**Impotance**

- Weights and biases collectively determine the behavior of the neural network during forward propagation.
- The network learns optimal values for weights and biases during training, adjusting them to minimize the difference between predicted and actual outputs (the loss function).
- The learned weights and biases encode the knowledge gained by the network during training, allowing it to make accurate predictions on new, unseen data.

# Q8. Can you explain the concept of the chain rule and its application in backward propagation?

Certainly! In neural networks, the chain rule is applied during backward propagation to find how changes in the model's parameters (weights and biases) influence the overall loss. The idea is to break down the derivative of the loss with respect to a parameter into the product of smaller derivatives.

# Q9. What are some common challenges or issues that can occur during backward propagation, and how can they be addressed?

Back Propogation is a usefull step while training of the data but apart from it's advanatages there are some disadvantages, which make it limited in some sorts but here we discussed these challanges and their solutions as well . 

- Because of repaeting the data through many layers , the gradient may become very small. Which making it challenging for early layers to learn effectively. We can use the activation functions that mitigate vanishing gradients, such as ReLU. Batch normalization and gradient clipping can also help stabilize training.

- The model may memorize training data, leading to poor generalization on unseen data.To solve this problem we can use the  regularization techniques such as dropout, L1 or L2 regularization, or early stopping. Increase the size of the training dataset or use data augmentation.

-  For very deep networks, memory requirements during backward propagation can be substantial.We can use techniques like gradient checkpointing, which trades computation for memory, or consider model parallelism.

- Choosing an inappropriate learning rate may result in slow convergence or instability.We can solve this problem by experiment with different learning rates or use adaptive learning rate methods. Techniques like learning rate schedules or learning rate annealing can be beneficial.
